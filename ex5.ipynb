```python
# ==============================
# 1. Imports and Configuration
# ==============================
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import (
    confusion_matrix,
    classification_report,
    roc_auc_score,
    roc_curve,
    precision_recall_curve,
    auc
)
from sklearn.utils.class_weight import compute_class_weight

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, BatchNormalization
from tensorflow.keras.optimizers import Adam

RANDOM_STATE = 42
np.random.seed(RANDOM_STATE)
tf.random.set_seed(RANDOM_STATE)

# ==============================
# 2. Load Dataset
# ==============================
url = "https://storage.googleapis.com/download.tensorflow.org/data/creditcard.csv"
df = pd.read_csv(url)

X = df.drop("Class", axis=1)
y = df["Class"]

print("Class distribution:")
print(y.value_counts())
print(y.value_counts(normalize=True))

# ==============================
# 3. Train-Test Split (80–20)
# ==============================
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    stratify=y,
    random_state=RANDOM_STATE
)

# ==============================
# 4. Feature Scaling
# ==============================
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# ==============================
# 5. Handle Class Imbalance (Class Weights)
# ==============================
class_weights = compute_class_weight(
    class_weight="balanced",
    classes=np.unique(y_train),
    y=y_train
)
class_weights = dict(enumerate(class_weights))

print("\nClass Weights:", class_weights)

# ==============================
# 6. Build Deep Neural Network
# ==============================
model = Sequential([
    Dense(128, activation='relu', input_shape=(X_train_scaled.shape[1],)),
    BatchNormalization(),
    Dropout(0.3),

    Dense(64, activation='relu'),
    BatchNormalization(),
    Dropout(0.3),

    Dense(32, activation='relu'),
    Dropout(0.2),

    Dense(1, activation='sigmoid')
])

model.compile(
    optimizer=Adam(learning_rate=0.001),
    loss='binary_crossentropy',
    metrics=['accuracy']
)

model.summary()

# ==============================
# 7. Train the Model
# ==============================
history = model.fit(
    X_train_scaled,
    y_train,
    epochs=30,
    batch_size=2048,
    class_weight=class_weights,
    verbose=1
)

# ==============================
# 8. Plot Training Loss Curve
# ==============================
plt.figure(figsize=(6,4))
plt.plot(history.history['loss'], label='Training Loss')
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("Training Loss Curve (Experiment 1)")
plt.legend()
plt.show()

# ==============================
# 9. Model Evaluation on Test Set
# ==============================
y_test_probs = model.predict(X_test_scaled).ravel()
y_test_preds = (y_test_probs >= 0.5).astype(int)

# Confusion Matrix
cm = confusion_matrix(y_test, y_test_preds)
plt.figure(figsize=(5,4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.title("Confusion Matrix (Experiment 1)")
plt.show()

# Classification Report
print("\nClassification Report (Experiment 1):")
print(classification_report(y_test, y_test_preds, digits=4))

# ==============================
# 10. ROC Curve and ROC-AUC
# ==============================
fpr, tpr, _ = roc_curve(y_test, y_test_probs)
roc_auc = roc_auc_score(y_test, y_test_probs)

plt.figure(figsize=(5,4))
plt.plot(fpr, tpr, label=f"ROC-AUC = {roc_auc:.4f}")
plt.plot([0,1], [0,1], '--')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve (Experiment 1)")
plt.legend()
plt.show()

# ==============================
# 11. Precision-Recall Curve and PR-AUC
# ==============================
precision, recall, _ = precision_recall_curve(y_test, y_test_probs)
pr_auc = auc(recall, precision)

plt.figure(figsize=(5,4))
plt.plot(recall, precision, label=f"PR-AUC = {pr_auc:.4f}")
plt.xlabel("Recall")
plt.ylabel("Precision")
plt.title("Precision-Recall Curve (Experiment 1)")
plt.legend()
plt.show()

# ==============================
# 12. Final Metrics Summary
# ==============================
print(f"ROC-AUC (Experiment 1): {roc_auc:.4f}")
print(f"PR-AUC  (Experiment 1): {pr_auc:.4f}")
```
